```python
import pandas as pd
```

```python
df_produtos = pd.read_csv("D:/Area de trabalho/Data-Science-Learning/02-Python/Base-de-dados/produtos.csv", sep=',' )
df_produtos.shape
```

```python
df_produtos.dtypes
```

```python
df_produtos.head()
```

```python
# Alterarando tipo numerico

df_produtos['Quantidade'] = pd.to_numeric(df_produtos['Quantidade'])
df_produtos['Preço '] = pd.to_numeric(df_produtos['Preço'])
```

```python
#1- Desempenho de vendas por categoria

vendas_categoria = df_produtos.groupby('Categoria')['Quantidade'].sum()
vendas_categoria
```

```python
#2 -Total de unidade vendidas por produto

produto_mais_vendidos = df_produtos.groupby('Produto')['Quantidade'].sum()
produto_mais_vendidos
```

```python
#3 - Produto Mais vendido
unidade_vendidas = df_produtos.groupby('Produto')['Quantidade'].sum().idxmax()
unidade_vendidas
```

```python
#4 - Preço Médio por Produto
media_preco_produto = df_produtos.groupby('Produto')['Preço'].mean()
media_preco_produto
```

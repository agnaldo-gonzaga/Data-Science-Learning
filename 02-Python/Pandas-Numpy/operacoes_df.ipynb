```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Cria um Dicionário com informações de colaboradores
data = {
    'Nome': ['Alice','Bob', 'Charlie','David', 'Emily'],
    'Idade': [25,30,35,40,27],
    'Cargo': ['Analista', 'Gerente','CEO','Analista','Cordenador'],
    'Salario':[5000, 8000, 15000, 4800, 600]
}

```

```python
df = pd.DataFrame(data)   
df
```

```python
df.head(2) #Pega as 2 primeiras linhas
```

```python
df.tail() #Pega as 5 ultimas linhas
```

```python
df.info()
```

```python
df.describe()
```

```python
df[df['Salario'] > 5000] # Mostra o salario maior que 5000 R$
```

```python
df.sort_values(by='Idade', ascending=False) # Ordena de forma Decrecente
```

```python
df['Bônus'] = df['Salario'] * 0.1 # Cria uma coluna com Bônus de 10% do salario
df
```

```python
df.groupby('Cargo')['Salario'].mean() # Faz um agrupamento por cargo e mostra a média do salario
```

```python
df.plot(
    kind='bar',
    x='Nome',
    y='Salario',
    title='Salário dos Funcionários',
    rot=45
)
```

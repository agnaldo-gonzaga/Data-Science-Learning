Este notebook registra o aprendizado de fundamentos da biblioteca **Pandas**

```python
import pandas as pd
```

```python
df = pd.read_json(r"D:\Area de trabalho\Data-Science-Learning\02-Python\Base-de-dados\receitas.json",encoding="utf-8")
df
```

```python
receitas_expandidas = []

for receita in df['Receita']:
    total_ingredientes = len(receita['Ingredientes'])
    receitas_info = {
        'Titulo': receita['Titulo'],
        'Tempo de Preparo': receita['Tempo de Preparo'],
        'Porções': receita['Porções'],
        'Total de Ingredientes': total_ingredientes
    }
    receitas_expandidas.append(receitas_info)

print(len(receitas_expandidas)) 
df_expandido = pd.DataFrame(receitas_expandidas)
df_expandido
```

## Dia 03 | SQL | Menor e maior cidade por comprimento de nome na tabela STATION

```python
### 📋 Desafio 1

-- Consulte as duas cidades em STATION com os nomes de CIDADES mais curto e mais longo, assim como seus respectivos comprimentos (ou seja: número de caracteres no nome). Se houver mais de uma cidade menor ou maior, escolha a que aparece primeiro quando ordenada alfabeticamente.


-- Cidade com nome mais CURTO (primeira alfabeticamente em caso de empate)
SELECT CITY, LENGTH(CITY)
FROM STATION
ORDER BY LENGTH(CITY) ASC, CITY ASC
LIMIT 1;

-- Cidade com nome mais LONGO (primeira alfabeticamente em caso de empate)
SELECT CITY, LENGTH(CITY)
FROM STATION
ORDER BY LENGTH(CITY) DESC, CITY ASC
LIMIT 1;
```

# #Dia 04 | SQL | Cidades sem duplicatas terminadas em vogais da tabela STATION

```python
### 📋 Desafio 

-- Consulte a lista de nomes de CIDADES terminando com vogais (a, e, i, o, u) do STATION. Seu resultado não pode conter duplicados.

SELECT DISTINCT CITY
FROM STATION
WHERE LOWER(RIGHT(CITY, 1)) IN ('a', 'e', 'i', 'o', 'u');
```

# #Dia 05 | SQL | Seleção de cidades que iniciam e terminam com vogais na tabela STATION

```python
### 📋 Desafio 01


-- Consulte a lista de nomes de CIDADES a partir de STATION que tenham vogais (ou seja, a, e, i, o e u) como primeiro e último caractere. Seu resultado não pode conter duplicados.

SELECT DISTINCT CITY
FROM STATION
WHERE LOWER(LEFT(CITY, 1)) IN ('a', 'e', 'i', 'o', 'u')
  AND LOWER(RIGHT(CITY, 1)) IN ('a', 'e', 'i', 'o', 'u');
```

# #Dia 06 | SQL | Seleção de cidade que não começam com vogais ou não terminam com vogais

```python
### 📋 Desafio 01

-- Consulte a lista de nomes de CIDADES em STATION que não começam com vogais ou não terminam com vogais. Seu resultado não pode conter duplicados.

SELECT DISTINCT CITY FROM STATION 
WHERE SUBSTR(CITY, 1, 1) NOT IN ('a','e','i','o','u','A','E','I','O','U') 
OR SUBSTR(CITY, LENGTH(CITY), 1) NOT IN ('a','e','i','o','u','A','E','I','O','U');
```

# #Dia 07 | SQL | nomes de CIDADES a partir de STATION que não começam com vogais nem terminam com vogais

```python
### 📋 Desafio 01

-- Consulte a lista de nomes de CIDADES a partir de STATION que não começam com vogais nem terminam com vogais. Seu resultado não pode conter duplicados.



SELECT DISTINCT CITY
FROM STATION
WHERE CITY NOT LIKE 'A%' AND CITY NOT LIKE 'E%'
  AND CITY NOT LIKE 'I%' AND CITY NOT LIKE 'O%'
  AND CITY NOT LIKE 'U%'
  AND CITY NOT LIKE 'a%' AND CITY NOT LIKE 'e%'
  AND CITY NOT LIKE 'i%' AND CITY NOT LIKE 'o%'
  AND CITY NOT LIKE 'u%'
  AND CITY NOT LIKE '%a' AND CITY NOT LIKE '%e'
  AND CITY NOT LIKE '%i' AND CITY NOT LIKE '%o'
  AND CITY NOT LIKE '%u' AND CITY NOT LIKE '%A'
  AND CITY NOT LIKE '%E' AND CITY NOT LIKE '%I'
  AND CITY NOT LIKE '%O' AND CITY NOT LIKE '%U';
```

## Dia 03 | SQL | Menor e maior cidade por comprimento de nome na tabela STATION

```python
### 📋 Desafio 1

-- Consulte as duas cidades em STATION com os nomes de CIDADES mais curto e mais longo, assim como seus respectivos comprimentos (ou seja: número de caracteres no nome). Se houver mais de uma cidade menor ou maior, escolha a que aparece primeiro quando ordenada alfabeticamente.


-- Cidade com nome mais CURTO (primeira alfabeticamente em caso de empate)
SELECT CITY, LENGTH(CITY)
FROM STATION
ORDER BY LENGTH(CITY) ASC, CITY ASC
LIMIT 1;

-- Cidade com nome mais LONGO (primeira alfabeticamente em caso de empate)
SELECT CITY, LENGTH(CITY)
FROM STATION
ORDER BY LENGTH(CITY) DESC, CITY ASC
LIMIT 1;
```

# Dia 10 | SQL | identificação do tipo de triângulo pelos três lados.

```python
### 📋 Desafio 01

-- Julia pediu aos alunos que criassem alguns desafios de programação. 
Escreva uma consulta para imprimir o hacker_id, o nome e o número total de desafios criados por cada aluno.
 Ordene seus resultados pelo número total de desafios em ordem decrescente. Se mais de um aluno criou o mesmo número de desafios,
  então ordene o resultado por hacker_id. Se mais de um aluno criou o mesmo número de desafios e a contagem for menor que o número máximo de desafios criados, 
  então exclua esses alunos do resultado.

SELECT 
    h.hacker_id,
    h.name,
    COUNT(c.challenge_id) AS total_challenges
FROM Hackers h
JOIN Challenges c
    ON h.hacker_id = c.hacker_id
GROUP BY h.hacker_id, h.name
HAVING 
    COUNT(c.challenge_id) = (
        SELECT MAX(cnt)
        FROM (
            SELECT COUNT(*) AS cnt
            FROM Challenges
            GROUP BY hacker_id
        ) x
    )
    OR COUNT(c.challenge_id) IN (
        SELECT cnt
        FROM (
            SELECT COUNT(*) AS cnt
            FROM Challenges
            GROUP BY hacker_id
        ) y
        GROUP BY cnt
        HAVING COUNT(cnt) = 1
    )
ORDER BY 
    total_challenges DESC,
    h.hacker_id;


```

# Dia 10 | SQL | identificação do tipo de triângulo pelos três lados.

```python
### 📋 Desafio 01

--  Escreva uma consulta identificando o tipo de registro na tabela TRIANGLES usando seus três lados. Produza uma das seguintes instruções para cada registro na tabela:
Equilátero: É um triângulo com lados de igual comprimento.
Isósceles: É um triângulo com lados de igual comprimento.
Escaleno: É um triângulo com lados de comprimentos diferentes.
Não é um triângulo: Os valores dados de A, B e C não formam um triângulo.

SELECT
    CASE
        WHEN A + B <= C OR A + C <= B OR B + C <= A THEN 'Not A Triangle'
        WHEN A = B AND B = C THEN 'Equilateral'
        WHEN A = B OR B = C OR A = C THEN 'Isosceles'
        ELSE 'Scalene'
    END AS triangle_type
FROM TRIANGLES;
```

# Dia 11 | SQL | formatação de nomes com inicial da ocupação e contagem de ocorrências por profissão

```python
### 📋 Desafio 01

Gerar os seguintes dois conjuntos de resultados:

Consulte uma lista ordenada alfabeticamente de todos os nomes em OCUPAÇÕES, imediatamente seguida pela primeira letra de cada profissão como parêntese (ou seja: entre parênteses). Por exemplo: , , , e . AnActorName(A)ADoctorName(D)AProfessorName(P)ASingerName(S)
Consulte o número de ocorrências de cada ocupação em OCUPAÇÕES. Ordene as ocorrências em ordem crescente e as produza no seguinte formato:

There are a total of [occupation_count] [occupation]s.
onde é o número de ocorrências de uma ocupação em OCUPAÇÕES e é o nome minúsculo da ocupação. Se mais de uma ocupação tiver a mesma forma, elas devem ser ordenadas alfabeticamente.[occupation_count][occupation][occupation_count]

Nota: Haverá pelo menos duas entradas na tabela para cada tipo de ocupação.


SELECT Name || '(' || SUBSTR(Occupation, 1, 1) || ')' AS result
FROM OCCUPATIONS
ORDER BY Name;


SELECT 'There are a total of ' || COUNT(*) || ' ' || LOWER(Occupation) || 's.'
FROM OCCUPATIONS
GROUP BY Occupation
ORDER BY COUNT(*), Occupation;

```

```python
# Amostragem aleatória simples | Parâmetros: dataset, quantidade de amostras
def amostragem_aleatoria_simples(dataset, amostras):
    return dataset.sample(n = amostras)
```

## Amostragem sistemática

```python
dataset.shape
```

```python
len(dataset) // 100 # Retorna o tamanho do Dataset / 100
```

```python
random.seed(1)
random.randint(0, 325) # Sempre retorna o mesmo número
```

```python
# Gera uma sequencia de numerosc comaçando em 68  e indo até o final do dataset e pula de 325 em 325
np.arange(68, len(dataset), step = 325) 
```

```python
# Amostragem sistemática | Seleciona linhas em intervalos regulares com início aleatório
def amostragem_sistematica(dataset, amostras):
    intervalo = len(dataset) // amostras
    random.seed(1)
    inicio = random.randint(0, intervalo)
    indices = np.arange(inicio, len(dataset),step = intervalo)
    amostragem_sistematica = dataset.iloc[indices]
    return amostragem_sistematica
```

```python
df_amostra_sistematica = amostragem_sistematica(dataset, 100 )
df_amostra_aleatorioa_simples.shape
```

## Amostragem por grupos

```python
len(dataset) / 10
```

```python
grupos = []
id_grupo = 0
contagem = 0
LIMITE_GRUPO = 3256  # constante nomeada

for _ in dataset.iterrows():
    grupos.append(id_grupo)
    contagem += 1
    if contagem > LIMITE_GRUPO:
        contagem = 0
        id_grupo += 1

# Resumo dos grupos criados
grupos_unicos, contagens = np.unique(grupos, return_counts=True)

print(f"Total de grupos criados: {id_grupo + 1}")
print(f" Linhas por grupo: {contagens}")
print(f" Total de linhas processadas: {len(grupos)}")
```

```python
# Verifica quantos grupos foram criados e quantas linhas cada um tem
np.unique(grupos, return_counts=True) 
```

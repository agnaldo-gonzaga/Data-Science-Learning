## Amostragem aleatória simples

```python
# Amostra aleatória simples de 100 registros do dataset
df_amostra_aleatorioa_simples = dataset.sample(n=100)
```

```python
# Amostragem aleatória simples | Parâmetros: dataset, quantidade de amostras
def amostragem_aleatoria_simples(dataset, amostras):
    return dataset.sample(n = amostras)
```

```python
df_amostra_aleatorioa_simples = amostragem_aleatoria_simples(dataset, 100)
df_amostra_aleatorioa_simples.shape
```
